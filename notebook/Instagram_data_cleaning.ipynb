```python
# Rename Columns
df.columns = [
    "rank",
    "username",
    "influence_score",
    "posts",
    "followers",
    "avg_likes",
    "engagement_rate_60_days",
    "new_post_avg_likes",
    "total_likes",
    "country"
]
```

```python
# Clean Country Column
df["country"] = df["country"].str.strip().str.title()
```

```python
# Convert K / M / B values to Pure Numbers
def convert_to_number(value):
    if isinstance(value, str):
        value = value.lower().strip()
        
        if value.endswith('k'):
            return float(value.replace('k', '')) * 1000
        elif value.endswith('m'):
            return float(value.replace('m', '')) * 1000000
        elif value.endswith('b'):
            return float(value.replace('b', '')) * 1000000000
        else:
            return float(value)
    
    return value
```

```python
# Apply function to all relevant columns
cols_to_convert = [
    "posts",
    "followers",
    "avg_likes",
    "new_post_avg_likes",
    "total_likes"
]

for col in cols_to_convert:
    df[col] = df[col].apply(convert_to_number)
```

```python
# Convert to Numeric Type
df['posts'] = df['posts'].astype(int)
df['followers'] = df['followers'].astype(int)
df['total_likes'] = df['total_likes'].astype(int)
```

```python
df.info()
```

```python
# Verify Conversion
df[cols_to_convert].head()
```

```python
# Remove % and Convert to Float
df["engagement_rate_60_days"] = (
    df["engagement_rate_60_days"]
    .astype(str)
    .str.replace("%", "", regex=False)
    .astype(float)
)
```

```python
#Fill Null values
df['engagement_rate_60_days']= df['engagement_rate_60_days'].fillna(0)
```
